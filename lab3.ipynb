### Anonymize 
Anonymize the income and credit card datasets. Use Generalization or Supression methods on postal code. 

```python
income2['DOB'] = income2['DOB'].apply(str)
income2['DOB']=(2023 - pd.to_numeric(income2.DOB.str[:4]))
income2
```

```python
bins=[0,30,60,100]
income2['DOB']=pd.cut(income2['DOB'],bins)
income2
```

```python
creditcard2['DOB']=(2023 - pd.to_numeric(creditcard2.DOB.str[:4]))
bins=[0,30,60,100]

creditcard2
```

```python
creditcard2['DOB']=pd.cut(creditcard2['DOB'],bins)
creditcard2
```

```python
bins1=[0,30000,60000,100000]
income2['postal_code']=pd.cut(income2['postal_code'],bins1)
income2
```

```python
print(creditcard2['postal_code'].dtype)
print(income2['postal_code'].dtype)
```

```python
bins2=[0,30000,60000,100000]

creditcard2['postal_code']=pd.cut(creditcard2['postal_code'],bins2)
creditcard2
```
